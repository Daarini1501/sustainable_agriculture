```python
# Step 1: Upload file
from google.colab import files
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Upload the file
uploaded = files.upload()

# Automatically get uploaded filename
filename = list(uploaded.keys())[0]
print("File uploaded:", filename)

# Step 2: Read CSV into DataFrame
df = pd.read_csv(io.BytesIO(uploaded[filename]))

# Step 3: Basic Exploratory Data Analysis (EDA)
print("\nðŸ”¹ First 5 rows:")
display(df.head())

print("\nðŸ”¹ Last 5 rows:")
display(df.tail())

print("\nðŸ”¹ DataFrame Info:")
df.info()

print("\nðŸ”¹ Summary Statistics:")
display(df.describe(include='all'))

print("\nðŸ”¹ Column Data Types:")
print(df.dtypes)

print("\nðŸ”¹ Shape of DataFrame (rows, columns):", df.shape)

print("\nðŸ”¹ Missing Values per Column:")
print(df.isnull().sum())

print("\nðŸ”¹ Duplicate Rows:", df.duplicated().sum())

print("\nðŸ”¹ Column Names:")
print(df.columns.tolist())

# Step 4: Quick Visuals
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()

# Pairplot for first few numeric columns (if dataset is large, limit it)
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
if len(numeric_cols) > 1:
    sns.pairplot(df[numeric_cols[:5]])
    plt.show()

# Correlation Heatmap
if len(numeric_cols) > 1:
    plt.figure(figsize=(10,6))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
    plt.title("Correlation Heatmap")
    plt.show()
```
